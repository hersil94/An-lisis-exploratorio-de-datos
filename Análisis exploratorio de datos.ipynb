```python
#1.10 Transformar el tamaño (Size) de todas las aplicaciones a MB.
# El conjunto de datos SI CONSIDERA las limpiezas previas
#valores únicos
print("valores únicos") #se muestran los valores únicos para saber que caracteres hay en el conjunto de datos
print(set(df.Size))
def limpiarKilo (k): #función para pasa de kylobytes a megabytes
  fin=k[-1]       #Ultimo caracter de cada string(la letra k, o M)
  if fin=="k":
    n=float(k.strip("k")) #se limpia el caracter k y se pasa a numero
    mega=n/1000     #se divide el valor numerico por 1000
    string= str(mega)+"MB" #se pasa a string y se agrega MB al final
  else:
    string=str(k)
  return string
def limpiarMega(k): #función para pasa de la notación M a MB para megabytes, no cambia el "número"
  fin=k[-1] #Ultimo caracter de cada string(la letra k, o M)
  if fin=="M":
    n=float(k.strip("M")) # se limpia el caracter m
    mega=n
    string= str(mega)+"MB" #se agrega MB al final
  else:
    string=str(k)
  return string

kilo=df.Size.apply(limpiarKilo) #se aplican las funciones de limpieza definidas previamente
enMB=kilo.apply(limpiarMega)
enMB
df["Size"]=enMB
print("")
print("El conjunto de datos, en MB, es:")
df.Size

#No se eliminan las filas con varies with device porque no se considera un valor erroneo
#Se considera como un valor no numérico, pero que no amerita borrar la fila completa

```

```python
#1.15 Mostrar las primeras 10 aplicaciones con más instalaciones.
# El conjunto de datos SI CONSIDERA las limpiezas previas

#valores únicos
set(df.Installs)

def limpiarMas(x): #se define una función para quitar los "+" y convertir en número entero, la columna de installs
  fin=x[-1]
  if fin=="+":
    x=x.strip("+")
    z=x.replace(",","")
    z=int(z)
  return z


df3=df.copy() #se hace una copia porque se van a cambiar los datos del df
df3["Installs"]=df3["Installs"].apply(limpiarMas) #se limpian los "+"
orden=df3.sort_values(by=["Installs"],ascending=False) #se ordena segun installs
#
TopInstalls=df3[df3["Installs"]!="Free"] #No se incluye la fila que tenga free.
segun_installs=TopInstalls.sort_values(by=["Installs"], ascending=False) #ordenar segun instalaciones
segun_installs.head(10) #llamar las primeras 10
#Las instalaciones mostradas es un número redondeado hacia abajo
```

```python
#1.16 Mostrar las categorías de mayor a menor por cantidad de aplicaciones.
# El conjunto de datos SI CONSIDERA las limpiezas previas

##Función para calcular moda en listas
def calcularModa(lista):
    contadorModal=0 #es la frecuencia del elemento con mas apariciones actualmente
    modas=[]
    
    for elemento in lista: #para cada elemento
        conteo=lista.count(elemento) #se cuenta cuantas veces aparece en la lista
        if conteo> contadorModal: #si la frecuencia del elemento actual supera al que hasta ese punto tenía mayor frecuencia
            contadorModal=conteo #contadorModal se sobreescribe con la nueva mayor frecuencia            
    for elemento in lista:#se cuentan las apariciones de nuevamente
        conteo=lista.count(elemento)
        if conteo==contadorModal and elemento not in modas: #si durante el conteo algún elemento tiene la misma frecuencia que el contadorModal
            modas.append(elemento) #ese elemento es la moda y se agrega a la lista de modas, es válida para modas únicas y no únicas
    return modas, contadorModal
apps=df_copy["App"]
cat=df_copy["Category"]
i=0
lista=[]
topCategorias=[] #se arman listas con el top para el gráfico posterior
topCategoriasFreq=[]
while i< len(apps):
  lista.append(cat[i]) #se agrega cada categoria a una lista, con repeticion segun frecuencia
  i=i+1
while len(lista)>0: #la lista se irá acortando cada iteración que se tome lo necesario de ella
  j=0
  par=calcularModa(lista) #se calcular la moda de la lista
  moda=par[0][0] #moda
  frecuencia=par[1] #frecuencia de la moda
  print("la categoría ", moda, " tiene ", frecuencia," aplicaciones")
  topCategorias.append(moda)
  topCategoriasFreq.append(frecuencia)
  while j<frecuencia:
    lista.remove(moda) #se elimina la moda de la lista para repetir el proceso con las siguientes "submodas"
    j=j+1
#La categoría 1.9 no se elimina, aunque podría, de todas formas es la última categoría, con una sola app
```

```python
#2.3.Graficar las "edades" (Content rating) por cantidad de aplicaciones de mayor a menor.
# El conjunto de datos SI CONSIDERA las limpiezas previas

contentR=df_copy["Content Rating"]
cat=df_copy["Category"]
i=0
listaContent=[]
topContentR=[] #se arman listas con el top para el gráfico posterior
topContentRFreq=[]
while i< len(contentR):
  listaContent.append(contentR[i]) #se agrega cada categoria a una lista, con repeticion segun frecuencia
  i=i+1
while len(listaContent)>0: #la lista se irá acortando cada iteración que se tome lo necesario de ella
  j=0
  par=calcularModa(listaContent) #se calcular la moda de la lista
  moda=par[0][0] #moda
  frecuencia=par[1] #frecuencia de la moda
  
  topContentR.append(moda)
  topContentRFreq.append(frecuencia)
  while j<frecuencia:
    listaContent.remove(moda) #se elimina la moda de la lista para repetir el proceso con las siguientes "submodas"
    j=j+1

# Se elimina el valor nan de las listas
topContentR.pop(-1)
topContentRFreq.pop(-1)
#print(topContentR)
#print(topContentRFreq)

plt.figure(figsize=(10,8)) #lienzo

plt.title("Aplicaciones por edad")
#nombre de los ejes
plt.xlabel("Clasificación por edad (Content Rating)")
plt.ylabel("Cantidad de aplicaciones por edad")
plt.bar(topContentR,topContentRFreq, width=0.5, color="c") # se grafica reduciendo el ancho estándar y en color cian
plt.show()

```

```python
#7 Realizar un gráfico de las cantidad de aplicaciones agrupadas por categoria.
# El conjunto de datos SI CONSIDERA las limpiezas previas

#print(topCategorias)
#print(topCategoriasFreq)
def invertirLista(x): #se define una función para invertir listas
  i=-1
  inversa=[]
  while -i<=len(x):
    inversa.append(x[i])
    i=i-1
  return inversa

x=invertirLista(topCategorias) #se invierten ambas listas porque barh por defecto, las grafica como "pirámide"
y=invertirLista(topCategoriasFreq)
#print(x)


plt.figure(figsize=(25,15)) #tamaño de lienzo
plt.title("Cantidad de aplicaciones por categoría") #titulo
#nombre de los ejes
plt.xlabel("Cantidad de aplicaciones por categoría") #nombre ejes
plt.ylabel("Categoría")
plt.xlim(0,2000)
plt.barh(x,y,color="m") #graficar horizontal con color magenta
plt.show()

# No se eleminó la categoría 1.9 porque no se sabe a qué se refiere o si es un error
# Sin embargo, en principio, este valor no afecta en nada a los demás, por lo que se puede mostrar la información sin problemas
```
